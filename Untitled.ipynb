```python
@app.route('/')
def index():
    # Fetch economic indicators for Nigeria
    url = f'https://api.tradingeconomics.com/historical-data/country/nigeria?c={API_KEY}'
    response = requests.get(url)
    data = response.json()

    # Extract relevant information
    indicators = []
    for item in data:
        indicator = {
            'indicator': item.get('indicator'),
            'value': item.get('value'),
            'date': item.get('date')
        }
        indicators.append(indicator)

    return render_template('index.html', indicators=indicators)
```

```python
app.run(debug=True)
```
